```python
cols2= [str(i) for i in range(len(df2.columns))]
cols3= [str(i) for i in range(len(df3.columns))]
cols4= [str(i) for i in range(len(df4.columns))]
```

```python
dfbaseline= df2.ix[:,['1','9','18','24', '26', '28']]
dfbaseline= dfbaseline.sort_values('9')
#print(dfbaseline)

dffadqn= df3.ix[:,['1','9','18','24', '26', '28']]
dffadqn= dffadqn.sort_values('9')
#print(dffadqn)

dfbatch= df4.ix[:,['1','5','18','24', '26', '28']]
dfbatch= dfbatch.sort_values('5')
#print(dfbatch)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

########################## sync base-line Min ##############################

fig, ax = plt.subplots()
labels = []
for key, grp in dfbaseline.groupby('1'):
    ax = grp.plot(ax=ax, kind='line', x='9', y=['24'])
    labels.append('w'+str(key))
lines, _ = ax.get_legend_handles_labels()
ax.set_title("sync baseline")
ax.set_xlabel("Network Update")
ax.set_ylim(200, 1400)
ax.set_ylabel("Convergence: Number of Episodes (min)")
ax.legend(lines, labels, bbox_to_anchor=(1.3, 1.025))
plt.show()

########################## sync FADQN Min ##############################

fig, ax = plt.subplots()
labels = []
for key, grp in dffadqn.groupby('1'):
    ax = grp.plot(ax=ax, kind='line', x='9', y=['24'])
    labels.append('w'+str(key))
lines, _ = ax.get_legend_handles_labels()
ax.set_title("sync FADQN")
ax.set_xlabel("Network Update")
ax.set_ylim(200, 1400)
ax.set_ylabel("Convergence: Number of Episodes (min)")
ax.legend(lines, labels, bbox_to_anchor=(1.3, 1.025))
plt.show()

########################## communication round sync base-line Min ##############################

fig, ax = plt.subplots()
labels = []
for key, grp in dfbaseline.groupby('1'):
    ax = grp.plot(ax=ax, kind='line', x='9', y=['28'])
    labels.append('w'+str(key))
lines, _ = ax.get_legend_handles_labels()
ax.set_title("sync baseline")
ax.set_xlabel("Network Update")
ax.set_ylabel("Communication Rounds")
ax.legend(lines, labels, bbox_to_anchor=(1.3, 1.025))
plt.show()

########################## communication round sync FADQN Min ##############################

fig, ax = plt.subplots()
labels = []
for key, grp in dffadqn.groupby('1'):
    ax = grp.plot(ax=ax, kind='line', x='9', y=['28'])
    labels.append('w'+str(key))
lines, _ = ax.get_legend_handles_labels()
ax.set_title("sync FADQN")
ax.set_ylabel("Communication Rounds")
#ax.set_ylim(200, 100000)
ax.set_xlabel("Network Update")
ax.legend(lines, labels, bbox_to_anchor=(1.3, 1.025))
plt.show()


########################## batch size experiments sync fadqn Min ##############################

fig, ax = plt.subplots()
labels = []
for key, grp in dfbatch.groupby('1'):
    ax = grp.plot(ax=ax, kind='line', x='5', y=['24'])
    labels.append('w'+str(key))
lines, _ = ax.get_legend_handles_labels()
ax.set_title("batch size sync fadqn")
#ax.set_xlim([16,32,64,96, 128])
plt.xticks([16,32,64,96,128], [16,32,64,96,128])
ax.autoscale(False)
ax.set_xlabel("batch size")
ax.set_ylabel("Convergence: Number of Episodes (min)")
ax.legend(lines, labels, bbox_to_anchor=(1.3, 1.025))
plt.show()
```
