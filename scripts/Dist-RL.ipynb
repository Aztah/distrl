```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

##########################Min ##############################
fig, ax = plt.subplots()
labels = []
for key, grp in sdf.groupby('1'):
    ax = grp.plot(ax=ax, kind='line', x='9', y=['23'])
    labels.append('w'+str(key))
lines, _ = ax.get_legend_handles_labels()
ax.set_xlabel("Network Update")
ax.set_ylabel("Convergence: Number of Episodes (min)")
ax.legend(lines, labels, bbox_to_anchor=(1.3, 1.025))
plt.show()

########################### max ############################
fig, ax = plt.subplots()
labels = []
for key, grp in sdf.groupby('1'):
    ax = grp.plot(ax=ax, kind='line', x='9', y=['25'])
    labels.append('w'+str(key))
lines, _ = ax.get_legend_handles_labels()
ax.set_xlabel("Network Update")
ax.set_ylabel("Convergence: Number of Episodes (max)")
ax.legend(lines, labels, bbox_to_anchor=(1.3, 1.025))
plt.show()
########################### avg ############################
fig, ax = plt.subplots()
labels = []
for key, grp in sdf.groupby('1'):
    ax = grp.plot(ax=ax, kind='line', x='9', y=['17'])
    labels.append('w'+str(key))
lines, _ = ax.get_legend_handles_labels()
ax.set_xlabel("Network Update")
ax.set_ylabel("Convergence: Number of Episodes (avg)")
ax.legend(lines, labels, bbox_to_anchor=(1.3, 1.025))
plt.show()
```
